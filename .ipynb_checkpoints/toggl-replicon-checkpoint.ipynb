```python
def createTogglProject(name):
    togglAuth = (togglApiKey, togglPassword)
    tURL = "https://api.track.toggl.com/api/v8/projects"
    tHeaders = {"Content-Type": "application/json"}
    tjson = {'project':{'name':name,'wid':wid}}
    togglResponse = requests.post(tURL, headers=tHeaders, json=tjson, auth=togglAuth)
    if b'already been taken' in togglResponse.content:
        print("Project already exists. Skipping project creation.")
        print("\n")
        return False
    else:
        pythonResponse = json.loads(togglResponse.content)
        print("Success")
        print("\n")
        return pythonResponse["data"]["id"]
    
    time.sleep(1)
    
    
```

```python
def createTogglTask(projectID, tname):
    togglAuth = (togglApiKey, togglPassword)
    tURL = "https://api.track.toggl.com/api/v8/tasks"
    tHeaders = {"Content-Type": "application/json"}
    tjson = {"task":{"name":tname,"pid":projectID}}
    togglResponse = requests.post(tURL, headers=tHeaders, json=tjson, auth=togglAuth)
    if b'already been taken' in togglResponse.content:
        print("     Task already exists...skipping")
        print("\n")
        return False
    else:
        pythonResponse = json.loads(togglResponse.content)
        print("    Success")
        print("\n")
        return pythonResponse["data"]["id"]
    
    time.sleep(1)
```

```python
def deleteTogglProject(pid):
    togglAuth = (togglApiKey, togglPassword)
    tURL = "https://api.track.toggl.com/api/v8/projects/" + str(pid)
    togglResponse = requests.delete(tURL, auth=togglAuth)
    
    if not b'null' in togglResponse.content:
        return "Success"
    else:
        return "ERROR: Could not delete project"
```

```python
def deleteTogglTask(tid):
    togglAuth = (togglApiKey, togglPassword)
    tURL = "https://api.track.toggl.com/api/v8/tasks/" + str(tid)
    togglResponse = requests.delete(tURL, auth=togglAuth)
    
    if not b'null' in togglResponse.content:
        return "Success"
    else:
        return "ERROR: Could not delete task"
```

```python
def replicon(rjson):

    
    repliconURL = 'https://na1.replicon.com/HEG/RemoteApi/RemoteApi.ashx/8.29.84/'
    
    repliconAuth = (rusername, rpassword)

    repliconHeaders = {
        "Host": "na1.replicon.com",
        "Content-Type": "application/json",
        "Content-Length": "126",
        "X-Replicon-Security-Context": "User"
    }
    

    repliconResponse = requests.post(repliconURL, headers=repliconHeaders, json=rjson, auth=repliconAuth)
    pythonResponse = json.loads(repliconResponse.content)
    
    return pythonResponse
```

```python
def getRepliconUserIdentity(user):
    
    data = replicon({"Action": "Query", "DomainType": "Replicon.Domain.User", "QueryType": "UserByLoginName", "Args": [user]})
    
    identity = data['Value'][0]['Identity']
    return identity
```

```python
def getRepliconProjectIdentityFromCode(code):
    data = replicon({
    "Action": "Query",
    "QueryType": "ProjectByCode",
    "DomainType": "Replicon.Project.Domain.Project",
    "Args": [
      code
    ],
    "Load": [
      {
        "Relationship": "Tasks",
        "Load": [
          {
            "Relationship": "ChildTasks",
            "Load": [
              {
                "Relationship": "ChildTasks",
                "Load": [
                  {
                    "Relationship": "ChildTasks"
                  }
                ]
              }
            ]
          }
        ]
      }
    ]
  })

    identity = data['Value']#[0]['Identity']
    
    return identity
```

```python
def getRepliconProjectIdentity(identity):
    data = replicon({
    "Action": "Query",
    "QueryType": "ProjectByIds",
    "DomainType": "Replicon.Project.Domain.Project",
    "Args": [
      [
        identity
      ]
    ]
  })

    identity = data
    
    return identity
```

```python
def getRepliconProjectsList(userID):
    data = replicon({
    "Action": "Query",
    "QueryType": "ExpenseProjectsByUser",
    "DomainType": "Replicon.Project.Domain.Project",
    "Args": [
      {
        "__type": "Replicon.Domain.User",
        "Identity": "131"
      }
    ]
  })
    return data['Value']
```

```python
def getRepliconTasksbyUserandProject(userID, projectID):
    data = replicon(
        {
    "Action": "Query",
    "QueryType": "AllOpenTasksByUserAndProject",
    "DomainType": "Replicon.Project.Domain.Task",
    "Args": [
      {
        "__type": "Replicon.Domain.User",
        "Identity": userID
      },
      {
        "__type": "Replicon.Project.Domain.Project",
        "Identity": projectID
      }
    ]
  }
    )
    
    if 'Exception' in data['Status']:
        print(data)
        return data
    else:
        return data['Value']
```

```python
def purge():
    
    print("Purging closed projects and tasks...")
    print("\n")
    
    i = 1
    
    file = open("repliconList.dat","r")
    filecontent = file.read()
    if filecontent != "":
        rProjectList = json.loads(filecontent)
    file.close()
    
    tProjectList = getTogglProjects(wid)
    
    numP = len(rProjectList) + 1 
    
    for p in tProjectList:
        progress = i / numP * 100  
        
        print("Searching for closed items - Project " + str(i) + " of " + str(numP) + " - " + str(round(progress, 2)) + "%")
        
        # Purge tasks first
        tasklist = getTogglTasks(p['id'])
        if tasklist != None:
            for t in tasklist:
                if not str(t['id']) in str(rProjectList):
                    print(str(t['id']))
                    print("\n")
                    print(t['id'], "task not found in Replicon...purging")
                    print(deleteTogglTask(t['id']))
        
        if not str(p['id']) in str(rProjectList):
            print(p['id'], "project not found in Replicon...purging")
            print(deleteTogglProject(p['id'])) 
        
        i += 1
        
    print("\n")
    print("2/3 Purging complete!")
    print("\n")
```

```python
def rename():
    
    print("Renaming projects and tasks...")
    print("\n")
    
    file = open("repliconList.dat","r")
    filecontent = file.read()
    if filecontent != "":
        rProjectList = json.loads(filecontent)
    file.close()
    
    tProjectList = getTogglProjects(wid)
    
    for p in tProjectList:
        tname = p['name']
        
        # Find matching rname
        for rp in rProjectList:          
            rname = rp['Properties']['Name']
            if rp['togglProjectID'] == p['id']:
                break
        if rname != tname:
            print(renameTogglProject(rp['togglProjectID'], rname))
            
    print("\n")
    print("3/3 Renaming complete!")
    print("\n")        
```

```python
def getPreviousIdsFromCode(code):
    file = open("repliconList.dat","r")
    filecontent = file.read()
    if filecontent != "":
        rProjectList = json.loads(filecontent)
    file.close()
    
    tProjectList = getTogglProjects(wid)
    
    for p in rProjectList:
        
        if code in p['Properties']['Name']:
            tID = p['togglProjectID']
            rID = p['Identity']
            break
    return tID, rID
```

```python
def syncall():
    
    start = time.time()
    
    print("Starting at: " + str(datetime.datetime.fromtimestamp(start)))
    
    create()
    purge()
    rename()
    
    end = time.time()
    print("\n")
    print("Finished syncronization at: " + str(datetime.datetime.fromtimestamp(end)))
    print("Time elapsed: ", (end - start), "s")
    print("\n")
    print("\n")
```
